Task 6 : Prediction using Decission Tree algorithm 

```python
#necessary imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
```

```python
#reading the dataset
iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
#info about the dataset
iris.info()
```

```python
#statical description of the dataset
iris.describe()
```

```python
#checking for null values
iris.isnull().sum() 
```

Model building and trainig 

```python
#splitting the dataset into features and target
X = iris.drop(columns=['Id', 'Species'])
y = iris['Species']
```

```python
#model fitting
dtree = DecisionTreeClassifier()
dtree.fit(X, y)
```

```python
print(f'accuracy_score: {accuracy_score(y, dtree.predict(X))}')
```

Visualize the decission tree 

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Assuming that 'dtree' is your trained decision tree model
plt.figure(figsize=(15,10))
tree.plot_tree(dtree, 
               feature_names=iris.columns[1:5],  
               filled=True)
plt.show()
```
