Task 2 : Prediction using unsupervised ML 

predict the number of clusters and visualize them on the iris dataset

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

Data visualization

```python
plt.figure(figsize=(12,4))
plt.scatter(iris['PetalLengthCm'], iris['PetalWidthCm'],iris['SepalLengthCm'],iris['SepalWidthCm'])
plt.title('The Iris Dataset')
plt.show()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
#Checking for missing values
iris.isnull().sum()
```

Splting and fiding the optimal k

```python
X = iris.iloc[:, [0, 1, 2, 3]].values
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Using kmeans with the optimal k

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

Visualize the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```
